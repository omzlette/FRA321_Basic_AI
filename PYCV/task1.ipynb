**Pixelate using Python**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import urllib
```

```python
def ImgData(img):
    return img.shape, img.size, img.dtype
```

```python
# Pixelate Image
def pixelate(img, scalew, scaleh):
    img = cv2.resize(img, (scalew, scaleh), interpolation=cv2.INTER_LINEAR)
    img = cv2.resize(img, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)
    return img
```

```python
# Plot Image
def plotimages(imglist, titlelist, rows=1, cols=2):
    plt.figure(figsize=(18, 10))
    for imgidx, title in zip(range(len(imglist)), titlelist):
        plt.subplot(rows, cols, imgidx+1)
        plt.axis('off')
        plt.text(imglist[imgidx].shape[0]/2, imglist[imgidx].shape[1], ImgData(imglist[imgidx]), fontsize=10, color='white', bbox=dict(facecolor='black', alpha=0.5), ha='center')
        plt.title(title)
        plt.imshow(cv2.cvtColor(imglist[imgidx], cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# Original Image
req = urllib.request.urlopen('https://raw.githubusercontent.com/omzlette/FRA321_Basic_AI/main/PYCV/img/ncg_1.jpg')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1)

# Results
imglist = [img, pixelate(img, 3000, 3000), pixelate(img, 1000, 1000), pixelate(img, 500, 500), pixelate(img, 100, 100), pixelate(img, 50, 50)]
titlelist = ['Original', '3000x3000', '1000x1000', '500x500', '100x100', '50x50']
plotimages(imglist, titlelist, 2, 3)
```
