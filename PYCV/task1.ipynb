**Pixelate using Python**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import urllib
```

```python
def ImgData(img):
    return img.shape, img.size, img.dtype
```

```python
# Pixelate Image
def pixelate(img, scalew, scaleh):
    img = cv2.resize(img, (scalew, scaleh), interpolation=cv2.INTER_LINEAR)
    img = cv2.resize(img, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)
    return img
```

```python
# Plot Image
def plotimages(imglist, titlelist, rows=1, cols=2):
    plt.figure(figsize=(40, 10))
    for imgidx, title in zip(range(len(imglist)), titlelist):
        plt.subplot(rows, cols, imgidx+1)
        plt.axis('off')
        plt.text(0, 0, ImgData(imglist[imgidx]), fontsize=24, color='white', bbox=dict(facecolor='black', alpha=0.5))
        plt.title(title)
        plt.imshow(cv2.cvtColor(imglist[imgidx], cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
# Original Image
req = urllib.request.urlopen('https://pbs.twimg.com/media/Eu9dF6xVEAMqg6P?format=jpg&name=900x900')
arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
img = cv2.imdecode(arr, -1) # 'Load it as it is'

# Results
imglist = [img, pixelate(img, 400, 400), pixelate(img, 200, 200), pixelate(img, 100, 100)]
titlelist = ['Original', 'Pixelate 400x400', 'Pixelate 200x200', 'Pixelate 100x100']
plotimages(imglist, titlelist, 1, 4)
```
